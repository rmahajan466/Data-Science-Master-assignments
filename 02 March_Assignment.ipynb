```python
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. 
It is particularly popular in data science and scientific computing for its ability to produce high-quality plots and graphs with
a relatively simple syntax. Matplotlib can be used in various environments, including Jupyter notebooks, web applications, and standalone
scripts.


Matplotlib is used for the following reasons:

-> Versatility: It supports a wide range of plots and can be used for simple line plots, bar charts, histograms, scatter plots, \
    3D plots, and more complex visualizations.

-> Customization: It offers extensive control over plot appearance, including colors, labels, scales, and fonts, allowing 
    for highly customized visualizations.

 -> Integration: It integrates well with other Python libraries such as NumPy, pandas, and SciPy, making it an essential tool in 
    the data analysis workflow.
 
-> Publication Quality: It can produce publication-quality figures in various formats, including PNG, PDF, SVG, and EPS.
 
-> Interactive: It can create interactive plots that can be embedded in applications and web interfaces using backends like Tkinter, 
    PyQt, and Jupyter notebooks.

    
Five Plots Using the Pyplot Module of Matplotlib
1) Line Plot
2) Bar Graph
3) Histogram
4) pie Chart
5) Scattered Plot
```

```python
#Question-02

A scatter plot is a type of data visualization that uses Cartesian coordinates to display values for two variables. 
It shows how much one variable is affected by another, providing a visual representation of the relationship between the variables. 
Each point on the scatter plot represents an observation in the dataset, with the position determined by the values of the two variables.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create scatter plot
plt.scatter(x, y)

# Add title and labels
plt.title('Scatter Plot of Generated Data')
plt.xlabel('X Values')
plt.ylabel('Y Values')

# Display plot
plt.show()
```

```python
#question-03

The subplot() function in Matplotlib is used to create multiple plots in a single figure, allowing for a grid of subplots within one overall figure. 
This is particularly useful for comparing different plots side by side or for organizing multiple visualizations in a clean and structured 
manner. Each subplot can be independently configured with its own set of axes, titles, and labels.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create figure and subplots
plt.figure(figsize=(10, 8))

# Line plot 1
plt.subplot(2, 2, 1)
plt.plot(x1, y1, marker='o')
plt.title('Line Plot 1')
plt.xlabel('X1')
plt.ylabel('Y1')

# Line plot 2
plt.subplot(2, 2, 2)
plt.plot(x2, y2, marker='s')
plt.title('Line Plot 2')
plt.xlabel('X2')
plt.ylabel('Y2')

# Line plot 3
plt.subplot(2, 2, 3)
plt.plot(x3, y3, marker='^')
plt.title('Line Plot 3')
plt.xlabel('X3')
plt.ylabel('Y3')

# Line plot 4
plt.subplot(2, 2, 4)
plt.plot(x4, y4, marker='d')
plt.title('Line Plot 4')
plt.xlabel('X4')
plt.ylabel('Y4')

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plots
plt.show()
```

```python
#question-04

A bar plot is a type of chart that presents categorical data with rectangular bars, where the lengths or heights of the bars are 
proportional to the values they represent. The bars can be plotted vertically or horizontally. Bar plots are used to compare different 
categories or groups and to show trends or patterns in the data.

Bar plots are used for several reasons:

-> Comparison: They allow for easy comparison of quantities across different categories.

-> Clarity: They provide a clear and straightforward way to visualize differences in data values.

-> Trend Identification: They help in identifying trends or patterns in data.

-> Categorical Data: They are ideal for representing categorical data and discrete variables.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.figure(figsize=(10, 5))

# Vertical bar plot
plt.subplot(1, 2, 1)
plt.bar(company, profit, color=['blue', 'green', 'red', 'purple'])
plt.title('Company Profits (Vertical Bar Plot)')
plt.xlabel('Company')
plt.ylabel('Profit in USD')

# Horizontal bar plot
plt.subplot(1, 2, 2)
plt.barh(company, profit, color=['blue', 'green', 'red', 'purple'])
plt.title('Company Profits (Horizontal Bar Plot)')
plt.xlabel('Profit in USD')
plt.ylabel('Company')

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plots
plt.show()
```

```python
#Question-05

A box plot (also known as a box-and-whisker plot) is a standardized way of displaying the distribution of data based on 
a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It may also show outliers.

Box plots are used for several reasons:

-> Summary Visualization: They provide a visual summary of the distribution of a dataset, including its central tendency, variability, 
    and skewness.

-> Comparative Analysis: They are useful for comparing distributions between different groups or datasets.

-> Outlier Detection: They help in identifying outliers in the data.

-> Distribution Shape: They give a clear indication of the shape of the data distribution, such as symmetry and the presence of tails.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create box plot
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot of Two Data Sets')
plt.xlabel('Data Set')
plt.ylabel('Values')

# Display the plot
plt.show()
```
