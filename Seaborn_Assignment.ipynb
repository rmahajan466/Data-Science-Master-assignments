```python
#Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
```

```python
1. Scatter Plot:
   - Use: Scatter plots are used to visualize the relationship between two numerical variables. They are helpful in identifying patterns, trends, and correlations between the variables.

2. Histogram:
   - Use: Histograms are useful for displaying the distribution of a single numerical variable. They help in understanding the frequency or density distribution of the data and identifying any underlying patterns such as skewness or multimodality.

3. Bar Plot:
   - Use: Bar plots are ideal for comparing categorical variables. They display the distribution of a categorical variable by depicting the frequency or aggregate of another numerical variable for each category. Bar plots are effective in visualizing comparisons, such as group-wise analysis or comparing different categories.

4. Box Plot (or Box-and-Whisker Plot):
   - Use: Box plots are commonly used for visualizing the distribution and statistical summary (such as median, quartiles, and outliers) of a numerical variable. They are particularly useful for detecting outliers and understanding the spread and central tendency of the data across different categories.

5. Heatmap:
   - Use: Heatmaps are graphical representations of data where values in a matrix are represented as colors. They are often used to visualize the correlation matrix of numerical variables, making it easy to identify patterns of correlation or association between variables. Heatmaps are also used in fields like biology for visualizing gene expression data and in business for displaying correlation matrices in financial analysis.
```

```python
#Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
#"timepoint" and y = "signal" for different events and regions.
#Note: timepoint, signal, event, and region are columns in the fmri dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Plot line plot for different events and regions
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region", markers=True, dashes=False)
plt.title("Signal over Time for Different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Plot two box plots
plt.figure(figsize=(12, 6))

# Box plot for age
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic_data, x="pclass", y="age")
plt.title("Age Distribution Across Passenger Classes")
plt.xlabel("Passenger Class")
plt.ylabel("Age")

# Box plot for fare
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic_data, x="pclass", y="fare")
plt.title("Fare Distribution Across Passenger Classes")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot histogram with hue parameter
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x="price", hue="cut", kde=True)
plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.legend(title="Cut")
plt.show()
```

```python
#Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix for the heatmap
flights_pivot = flights_data.pivot("month", "year", "passengers")

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Passenger Numbers by Month and Year")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
